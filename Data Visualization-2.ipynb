```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array(['A','B','C','D','E','F'])
y=np.array([12,21,12,4,7,3])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['AA','A','AAA','CR7','Ever','NQE'])
y=np.array([50,10,72,110,27,200])
plt.bar(x,y)
plt.xlabel('Batteries')
plt.ylabel('Discharging time(hrs)')
plt.title('performance')
plt.show()
```

```python
x=np.array(['AA','A','AAA','CR7','Ever','NQE'])
y=np.array([50,10,72,110,27,200])
plt.bar(x,y)
plt.xlabel('Batteries')
plt.ylabel('Discharging time(hrs)')
plt.title('performance',color='r')
plt.show()
```

```python
x=np.array(['AA','A','AAA','CR7','Ever','NQE'])
y=np.array([50,10,72,110,27,200])
plt.barh(x,y)
plt.xlabel('Batteries')
plt.ylabel('Discharging time(hrs)')
plt.title('performance',color='r')
plt.show()
```

```python
x=np.array(['AA','A','AAA','CR7','Ever','NQE'])
y=np.array([50,10,72,110,27,200])
plt.barh(x,y,color='g')#bar to green
plt.xlabel('Batteries')
plt.ylabel('Discharging time(hrs)')
plt.title('performance',color='r')
plt.show()
```

```python
x=np.random.normal(170,10,250)#gaussian data distribution
plt.hist(x)
plt.show()
```

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['Audi','BYD','Kia','Volkswagen']
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['Audi','BYD','Kia','Volkswagen']
myexplode=[0,0.12,0.07,0.05]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```
